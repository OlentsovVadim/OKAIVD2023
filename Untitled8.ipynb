```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import seaborn as sns
from sklearn.impute import KNNImputer
from sklearn.decomposition import PCA
from sklearn.preprocessing import MinMaxScaler, StandardScaler
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
from IPython.core.display import display, HTML
```

```python
file_raw = pd.read_csv("https://raw.githubusercontent.com/EvgeniiSopovGitHub/datasets_uni/main/okaivd/lab_4_raw_data.csv", sep = ';' , index_col=[0])
columns = file_raw.columns
data=file_raw.index
```

```python
file_raw.info()
```

```python
imputer = KNNImputer(n_neighbors = 3)
rez = imputer.fit_transform(file_raw)
file_2 = pd.DataFrame(rez, columns=columns, index=data)
file_2.head()
```

```python
file_3 =file_2.copy()
file_3
file_3['min']=file_2.min(axis=1)
file_3['median']=file_2.median(axis=1)
file_3['max']=file_2.max(axis=1)

file_3['mean/std']=round(file_2.mean( axis = 1), 2).astype( str ) + '/' + round(file_2.std(axis = 1), 2).astype(str)
mean=file_3.mean(axis=0)
file_3=file_3.T

file_3['average annual']=mean
file_4=file_3.T
file_4.to_csv('lab4_bss22-04_Bazin', index=None)
file_4
```
